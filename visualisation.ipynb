```python
# importons les bibliothèques que nous avons utiliser  
import pandas as pd 
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv(r"/content/website_count.csv")
```

```python
#convertir le fichier CSV en liste
list_of_csv = [list(row) for row in df.values]
```

```python
#supprimer les []
liste=[]
for i in range(len(list_of_csv)):
    string =list_of_csv[i]
    str = ' '.join(string)
    liste.append(str)
    
```

```python
#Supprimer certains caracteres d une chaine de caracteres en utilisant .replace()
characters = "'([)]"
liste1 = []
for i in range(len(list_of_csv)):
    string =liste[i]
    for x in range(len(characters)):
        string= string.replace(characters[x],"")
    liste1.append(string)
```

```python
liste1
```

```python
# Divisez la chaîne en utilisant une virgule suivie d'un espace comme séparateur 
liste3=[]
for i in range(len(liste1)):
    x = liste1[i].split(", ")
    liste3.append(x)
    
    
```

```python
liste3
```

```python
#convertir une liste en CSV 
cols = ["site","nbr"]
with open('akka.csv', 'w') as f:
    # using csv.writer method from CSV package
    write = csv.writer(f)
    write.writerow(cols)
    write.writerows(liste3)
```

```python
dd=pd.read_csv(r"/content/akka.csv")
dd
```

```python
# Appliquez maintenant la fonction groupby() et la metode max()
da=dd.groupby(['site']).max()
```

```python
da
```

```python
da.dtypes
```

```python
# changer l index
da.reset_index(inplace=True)
```

```python
da
```

```python

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = da["site"]
sizes = da["nbr"]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
